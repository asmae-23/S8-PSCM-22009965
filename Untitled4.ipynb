```python
# 1. Installation de la bibliothèque nécessaire
!pip install qrcode[pil]

import qrcode
from PIL import Image

# 2. Configure ici le lien ou le texte que tu veux transformer en QR Code
# Remplace l'URL ci-dessous par le lien de ton projet ou de ton dataset
data = "https://www.kaggle.com/datasets/juzersharif/supply-chain-shipment-dataset"

# 3. Création du QR Code
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)
qr.add_data(data)
qr.make(fit=True)

# 4. Génération de l'image
img = qr.make_image(fill_color="black", back_color="white")

# 5. Sauvegarde et affichage
img.save("qrcode_projet_supply_chain.png")
print("Le QR Code a été généré avec succès !")
img
```
